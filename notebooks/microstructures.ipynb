```python
class SigmaClastData:
    IMG_SIZE = 1000
    CW = './original_data/CW'
    CCW = './original_data/CCW'
    OUTPUT_DIR = Path('./processed_data/')
    LABELS = {CW: 0, CCW: 1}
    cw_count = 0
    ccw_count = 0
    
    def __init__(self, rebuild_data=False):
        self.REBUILD_DATA = rebuild_data  # GETS SET TO FALSE AFTER FIRST TIME RUNNING
    
    def make_training_data(self):
        if not self.OUTPUT_DIR.exists():
            self.OUTPUT_DIR.mkdir(parents=True)
            
        for label, value in self.LABELS.items():
            for f in tqdm(Path(label).iterdir()):
                try:
                    img = cv2.imread(str(f), cv2.IMREAD_GRAYSCALE)
                    img = cv2.resize(img, (self.IMG_SIZE, self.IMG_SIZE))

                    flipped_img = cv2.flip(img, 1)  # flips image horizonally

                    # zip img together with 1-hot label vector
                    img = np.array((np.array(img), np.eye(2)[value]), dtype=object)

                    if label == self.CW:
                        flipped_img = np.array(
                            (np.array(flipped_img), np.eye(2)[value]),
                            dtype=object
                        )
                        np.save(self.OUTPUT_DIR/f'CW_{self.cw_count:03n}.npy', img)
                        np.save(
                            self.OUTPUT_DIR/f'CW_{self.cw_count:03n}_flipped.npy',
                            flipped_img
                        )
                        self.cw_count += 1
                    else:
                        flipped_img = np.array(
                            (np.array(flipped_img), np.eye(2)[value]),
                            dtype=object
                        )
                        np.save(self.OUTPUT_DIR/f'CCW_{self.ccw_count:03n}.npy', img)
                        np.save(
                            self.OUTPUT_DIR/f'CCW_{self.ccw_count:03n}_flipped.npy',
                            flipped_img
                        )
                        self.ccw_count += 1
                except Exception as e:
                    print(e)
                    print(f'\n\nImg file {f} had an error\n\n')
                    continue
                
                self.REBUILD_DATA = False
                    
                
if REBUILD_DATA:
    sc = SigmaClastData()
    sc.make_training_data()
```
